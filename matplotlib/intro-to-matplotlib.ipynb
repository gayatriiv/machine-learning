```python
#import the library

%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare the data
x = [1,2,3,4]
y=[11,22,33,44]

# 2. setup plot
fig, ax = plt.subplots(figsize = (10,10))

# 3. plot data
ax.plot(x,y)

#4. customize plot
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

#5. save and show
fig.savefig("images/sample-plot.png")
```

# using numpy arrays for figures

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
x[:10]
```

```python
#line plot

fig,ax= plt.subplots()
ax.plot(x, x**2);
plt.show()
```

```python
# scatter plot

fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
plt.show();
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
plt.show();
```

```python
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots();
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
plt.show();
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
plt.show();
```
